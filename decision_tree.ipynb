```python
from sklearn.datasets import load_iris
iris = load_iris() # Call the load_iris function to get the Bunch object
x = iris.data
y = iris.target
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier # Corrected the typo in the class name
model = DecisionTreeClassifier(criterion='entropy') # Corrected the typo in the class name
model = model.fit(x, y)
# op.model.predict([[5,9,2,3,7,9,6]]) # This line may cause an error due to incorrect variable usage and shape mismatch
# print(op)
```

```python
fn=iris.feature_names
cn=iris.target_names
import graphviz
op=tree.export_graphviz(model,feature_names=fn,class_names=cn,filled=True,rounded=True,special_characters=True) # Changed true to True for boolean arguments
# The 'source' function should be 'Source' (capital 'S')
graph=graphviz.Source(op)
graph
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,train_size=0.7)
model=model.fit(xtrain,ytrain)
op=model.predict(xtest)
print(op)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,op)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(ytest,op)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
display = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
display.plot()
```
